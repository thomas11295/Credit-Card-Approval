# Machine Learning Approach

```python
x = df.drop(['label', 'Ind_ID'], axis=1) # defining x and droping label and ind id column bcoz they can not be useful
y = df['label'] # defining y the target variable
```

```python
y.value_counts() # checking balance of data
```

```python
from imblearn.over_sampling import SMOTE # treating imbalance in dataset with the help of smote function from imblearn over sampling library
smote = SMOTE(sampling_strategy='minority')
x_sm,y_sm = smote.fit_resample(x,y)

y_sm.value_counts() # data is balanced now
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_sm,y_sm,test_size=0.2,random_state=42,stratify=y_sm) # splitting data to feed in model
```

```python
y_train.value_counts() # checking balance
```

```python
y_test.value_counts() # checking balance
```

# Scalig data

```python
from sklearn.preprocessing import MinMaxScaler

column_scale = ['Annual_income', 'Age', 'Work_experince'] # scaling data with min max scaler so the model easily undrstand the data ans also increase performance

scaler = MinMaxScaler()

# Fit and transform the training set
x_train[column_scale] = scaler.fit_transform(x_train[column_scale])

# Transform the testing set using the parameters from the training set
x_test[column_scale] = scaler.transform(x_test[column_scale])
```

# Running Different Classifictaion Models

```python

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score,classification_report


# Decision Tree Classifier
DT = DecisionTreeClassifier(random_state=42)
DT.fit(x_train, y_train)
y_pred_dt = DT.predict(x_test)
score_dt = accuracy_score(y_test, y_pred_dt)
report_dt = classification_report(y_test, y_pred_dt)
print("Decision Tree Classifier:")
print(f'Accuracy: {score_dt}')
print(report_dt)



from sklearn.linear_model import LinearRegression

# Linear Regression Classifier
LR = LinearRegression()
LR.fit(x_train, y_train)
y_pred_lr = LR.predict(x_test)
y_pred_binary_lr = [1 if pred >= 0.5 else 0 for pred in y_pred_lr]
score_lr = accuracy_score(y_test, y_pred_binary_lr)
report_lr = classification_report(y_test, y_pred_binary_lr)
print("\nLinear Regression:")
print(f'Accuracy: {score_lr}')
print(report_lr)


# XGBoost Classifier
from xgboost import XGBClassifier

xg = XGBClassifier(random_state=42)
xg.fit(x_train,y_train)
y_pred_xg = xg.predict(x_test)
score_xg = accuracy_score(y_test,y_pred_xg)
report_xg = classification_report(y_test,y_pred_xg)
print("\nXG boost:")
print(f'Accuracy: {score_xg}')
print(report_xg)


# RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier

RF = RandomForestClassifier(n_estimators=100,random_state=42)
RF.fit(x_train,y_train)
y_pred_rf = RF.predict(x_test)
score_rf = accuracy_score(y_test,y_pred_rf)
report_rf = classification_report(y_test,y_pred_rf)
print("\nRandom forest:")
print(f'Accuracy: {score_rf}')
print(report_rf)
```

```python
import matplotlib.pyplot as plt

# Accuracy scores for each model
accuracies = [score_dt, score_lr, score_xg, score_rf]
models = ['Decision Tree', 'Linear Regression', 'XGBoost', 'Random Forest']

# Bar plot
plt.bar(models, accuracies, color=['blue', 'green', 'orange', 'red'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Model Comparison - Accuracy Scores')
plt.ylim(0, 1)  # Set y-axis limits between 0 and 1 for accuracy scores
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV

# Define parameter grids for XGBoost and RandomForest
param_grid_xg = {
    'n_estimators': [50, 100, 150],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.3]
}

param_grid_rf = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10]
}

# Perform Grid Search with cross-validation for XGBoost
grid_search_xg = GridSearchCV(XGBClassifier(random_state=42), param_grid=param_grid_xg, cv=3)
grid_search_xg.fit(x_train, y_train)

# Perform Grid Search with cross-validation for RandomForest
grid_search_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid=param_grid_rf, cv=3)
grid_search_rf.fit(x_train, y_train)

# Get the best parameters and models
best_params_xg = grid_search_xg.best_params_
best_model_xg = grid_search_xg.best_estimator_

best_params_rf = grid_search_rf.best_params_
best_model_rf = grid_search_rf.best_estimator_

# Evaluate the best models on the test set
y_pred_best_xg = best_model_xg.predict(x_test)
score_best_xg = accuracy_score(y_test, y_pred_best_xg)
report_best_xg = classification_report(y_test, y_pred_best_xg)

y_pred_best_rf = best_model_rf.predict(x_test)
score_best_rf = accuracy_score(y_test, y_pred_best_rf)
report_best_rf = classification_report(y_test, y_pred_best_rf)

# Compare the performance of the best models
print("Best XGBoost Model:")
print(f'Best Parameters: {best_params_xg}')
print(f'Accuracy: {score_best_xg}')
print(report_best_xg)

print("\nBest RandomForest Model:")
print(f'Best Parameters: {best_params_rf}')
print(f'Accuracy: {score_best_rf}')
print(report_best_rf)
```

```python
import matplotlib.pyplot as plt

# Accuracy scores for the best models
accuracies_best = [score_best_xg, score_best_rf]
models_best = ['Best XGBoost', 'Best RandomForest']

# Bar plot
plt.bar(models_best, accuracies_best, color=['green', 'blue'])
plt.xlabel('Model')
plt.ylabel('Accuracy')
plt.title('Comparison of Best Models - Accuracy Scores')
plt.ylim(0, 1)  # Set y-axis limits between 0 and 1 for accuracy scores
plt.show()
```
